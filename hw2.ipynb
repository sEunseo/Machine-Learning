```python
import numpy as np
import cv2    # k-means 함수 내장
import matplotlib.pyplot as plt   # 결과 그리기 위해 로딩
from sklearn.cluster import KMeans
```

```python
%matplotlib inline
```

```python
# Read in the image -> "/ 255 를 나눠서 컬러 RGB 컬러로 나오도록 함"
image1= cv2.imread('images/Airplane.jpg') / 255
image2 = cv2.imread('images/Tiger.jpg') / 255


#Airplane 올리기
plt.imshow(image1)
```

```python
#Tiger 올리기
plt.imshow(image2)
```

```python
# reshape Image

# reshape Airplane image
AirImage = image1.reshape(image1.shape[0]* image1.shape[1], image1.shape[2])
AirImage.shape

#reshape Tiger image
TigImage = image2.reshape(image2.shape[0] * image2.shape[1], image2.shape[2])
TigImage.shape
```

```python
# 군집화 모듈 KMeans를 fit()할 때의 오류 방지
!pip install threadpoolctl==3.1.0
!pip install numpy==1.21.4
```

```python
# import Kmeans library to Airplaie
airKM = KMeans(n_clusters = 2, random_state=40)
airKM.fit(AirImage)

# import Kmeans library to Tiger
tigKM = KMeans(n_clusters = 2, random_state=40)
tigKM.fit(TigImage)
```

```python
airShow = airKM.cluster_centers_[airKM.labels_] #Apply K-means to Airplane
tigShow = tigKM.cluster_centers_[tigKM.labels_] #Apply K-means to Tiger

#Reshape Image
#reshape Airplane images to show
airRes = airShow.reshape(image1.shape[0], image1.shape[1], image1.shape[2]) 
#reshape Tiger images to show
tigRes = tigShow.reshape(image2.shape[0], image2.shape[1], image2.shape[2]) 
```

```python
plt.imshow(airRes) #result of Airplane
```

```python
plt.imshow(tigRes) #result of Tiger
```
